```python
import pandas as pd

# Load your events CSV
df = pd.read_csv("hyderabad_events.csv")

# Create empty indicator (dummy) columns
df["sports_event"] = (df["event_type"] == "sports").astype(int)
df["concert"] = (df["event_type"] == "concert").astype(int)
df["public_holiday"] = (df["event_type"] == "holiday").astype(int)
df["public_demonstration"] = (df["event_type"] == "demonstration").astype(int)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
```
